### Multiple Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mtcars.csv")
```

```python
df
```

```python
df.corrwith(df['mpg'])
```

```python
salary = pd.read_csv("Salary_Data.csv")
salary.corr()
#if attributes are highly correlated then it is used for linear regression model
```

#### separate input and output

```python
x = df[['disp','hp','wt']]
y = df['mpg']
x
```

#### import the class

```python
from sklearn.linear_model import LinearRegression
```

#### create the object

```python
reg = LinearRegression()
```

#### train the model

```python
reg.fit(x,y)
```

#### predict the model

```python
disp = 221
hp = 102
wt = 2.91
reg.predict([[disp,hp,wt]])
```

```python
disp = 221
hp = 115
wt = 2.51
reg.predict([[disp,hp,wt]])
```

```python
# to show accuracy 
reg.score(x,y)

# machine learning accuray threshold value is atleast 80%
```

#### Decision tree classification

#### import the dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.shape
```

```python
df # 0 is purchased and 1 not purchased
```

```python
df.sample(5)
```

### separate input and output variables

```python
x = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

```python
x
```

```python
y.value_counts()
```

### visualize the data

```python
plt.title('Purchase Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'],x['EstimatedSalary'])
```

```python
plt.title('Purchase Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'],x['EstimatedSalary'],c = y) #color wise distribute values of y variable
```

#### import the class

```python
from sklearn.tree import DecisionTreeClassifier
```

#### create the object

```python
classifier = DecisionTreeClassifier(random_state = 0)
```

#### train the model

```python
classifier.fit(x,y)
```

#### plot the tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(64,36))
plot_tree(classifier,fontsize=15,feature_names=['age','sal'],class_names=['y','n'],filled=True,rounded=True);
```

#### predict on new data

```python
new = pd.DataFrame({
    'Age':[23,45,54,67],
    'EstimatedSalary':[45000,89000,23000,134000]
})
```

```python
new
```

```python
classifier.predict(new) #conclusion is first two will purchased and next two will not
```
