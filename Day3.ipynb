#### separate input and output variables

```python
# y is always variable 
x = df['YearsExperience'].values #values -> to convert series into array format
y = df['Salary'] # here y is in series format
```

```python
x
```

```python
y;
```

```python
x = x.reshape(30,1) # it is not inplace function again we need to store it
x
```

#### import the class 

```python
from sklearn.linear_model import LinearRegression
```

#### create the object of linear regression

```python
regressor = LinearRegression()
```

#### Train the algrithm with data

```python
regressor.fit(x,y)
```
